```python
import numpy as np
import matplotlib.pyplot as plt

# Define the topics and scores
topics = ['Topic 1', 'Topic 2', 'Topic 3', 'Topic 4', 'Topic 5']
scores = [80, 70, 85, 90, 75]  # Performance scores out of 100
average_scores = [60, 65, 70, 75, 65]  # Average scores for comparison

# Convert data to match the radar chart format
angles = np.linspace(0, 2 * np.pi, len(topics), endpoint=False).tolist()
scores += scores[:1]  # Close the loop
average_scores += average_scores[:1]  # Close the loop
angles += angles[:1]  # Close the loop

# Plot the radar chart
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angles, scores, color='blue', alpha=0.25, label='Student')
ax.fill(angles, average_scores, color='red', alpha=0.25, label='Average')

# Add topic labels
ax.set_yticks([20, 40, 60, 80, 100])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(topics)

# Add title and legend
ax.set_title('Performance per Topic')
ax.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1))

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define topics and scores
topics = ['Algebra', 'Geometry', 'Trigonometry', 'Statistics', 'Calculus']
student_scores = [60, 75, 65, 70, 80]  # Student's performance per topic
passing_threshold = [50, 50, 50, 50, 50]  # Minimum passing marks for progression

# Prepare data for the radar chart
angles = np.linspace(0, 2 * np.pi, len(topics), endpoint=False).tolist()
student_scores += student_scores[:1]  # Close the loop
passing_threshold += passing_threshold[:1]  # Close the loop
angles += angles[:1]  # Close the loop

# Plot radar chart
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angles, student_scores, color='blue', alpha=0.25, label='Student')
ax.plot(angles, passing_threshold, color='red', linestyle='dashed', label='Passing Threshold')

# Add topic labels
ax.set_yticks([20, 40, 60, 80, 100])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(topics)

# Add title and legend
ax.set_title('Grade Progression Assessment', va='bottom')
ax.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1))

plt.show()
```
