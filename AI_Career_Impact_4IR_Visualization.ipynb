```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Step 1: Load Data
data = {
    'Year': [2018, 2019, 2020, 2021, 2022],
    '4IR_Enrollment': [10000, 12000, 15000, 18000, 20000]
}
df = pd.DataFrame(data)

# Step 2: EDA - Visualize Trends
plt.plot(df['Year'], df['4IR_Enrollment'], marker='o')
plt.title('4IR Enrollment Trends')
plt.xlabel('Year')
plt.ylabel('Enrollment')
plt.grid()
plt.show()

# Step 3: Predict Future Enrollment
X = df[['Year']]
y = df['4IR_Enrollment']

model = LinearRegression()
model.fit(X, y)

# Predict for 2023-2025
future_years = np.array([2023, 2024, 2025]).reshape(-1, 1)
predictions = model.predict(future_years)

print("Predicted Enrollment for 2023-2025:", predictions)

# Step 4: Insights
for year, pred in zip([2023, 2024, 2025], predictions):
    print(f"Year {year}: Predicted Enrollment = {int(pred)}")
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Step 1: Hardcoded Data
data = {
    'Year': [2018, 2019, 2020, 2021, 2022],
    '4IR_Enrollment': [10000, 12000, 15000, 18000, 20000]
}

# Tuition Revenue per Student and Cost Assumptions
tuition_per_student = 5000  # Example: $5000 per student per year
fixed_costs = 1_000_000  # Example: $1M annually for infrastructure, salaries
variable_cost_per_student = 2000  # Example: $2000 per student for materials, utilities

# Step 2: Convert data to numpy arrays
years = np.array(data['Year']).reshape(-1, 1)
enrollments = np.array(data['4IR_Enrollment'])

# Step 3: EDA - Visualize Enrollment Trends
plt.figure(figsize=(10, 6))
plt.plot(data['Year'], data['4IR_Enrollment'], marker='o', linestyle='-', color='b')
plt.title('4IR Enrollment Trends (2018-2022)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Enrollment', fontsize=12)
plt.grid(True)
plt.show()

# Step 4: Predict Future Enrollment
model = LinearRegression()
model.fit(years, enrollments)

future_years = np.array([2023, 2024, 2025]).reshape(-1, 1)
predictions = model.predict(future_years)

# Step 5: Revenue and Profit Analysis
projected_enrollments = np.concatenate((enrollments, predictions))
revenue = projected_enrollments * tuition_per_student
variable_costs = projected_enrollments * variable_cost_per_student
profit = revenue - (fixed_costs + variable_costs)

# Visualize Predicted Enrollment vs. Revenue and Profit
years_extended = np.concatenate((data['Year'], [2023, 2024, 2025]))

plt.figure(figsize=(12, 8))
plt.plot(years_extended, projected_enrollments, marker='o', label='Enrollment', color='b')
plt.plot(years_extended, revenue / 1000, marker='o', label='Revenue (in $1000s)', color='g')
plt.plot(years_extended, profit / 1000, marker='o', label='Profit (in $1000s)', color='r')
plt.title('Projected Enrollment, Revenue, and Profit (2018-2025)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Values (in $1000s)', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()

# Step 6: Display Insights
print("Predicted Enrollment for 2023-2025:")
for year, pred in zip([2023, 2024, 2025], predictions):
    print(f"Year {year}: Predicted Enrollment = {int(pred)}")

print("\nProfit Analysis:")
for year, prof in zip(years_extended, profit):
    print(f"Year {int(year)}: Profit = ${int(prof):,}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Define Fields and Potential Impact
fields = [
    'Business',
    'Healthcare',
    'Engineering',
    'Education',
    'Government'
]

impact_metrics = {
    'Automation Potential': [80, 60, 70, 40, 50],
    'Data-Driven Decisions': [90, 70, 75, 50, 85],
    'Efficiency Gains': [85, 65, 80, 45, 55],
    'Innovation': [75, 80, 85, 60, 70]
}

# Step 2: Visualization - Heatmap of Data Science Impact
fig, ax = plt.subplots(figsize=(10, 7))
impact_values = np.array(list(impact_metrics.values()))

im = ax.imshow(impact_values, cmap='Blues', aspect='auto')

# Add labels for fields and metrics
ax.set_xticks(np.arange(len(fields)))
ax.set_yticks(np.arange(len(impact_metrics)))
ax.set_xticklabels(fields)
ax.set_yticklabels(impact_metrics.keys())

# Rotate the tick labels and set their alignment
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

# Loop over data dimensions and create text annotations
for i in range(len(impact_metrics)):
    for j in range(len(fields)):
        text = ax.text(j, i, f"{impact_values[i, j]}%", ha="center", va="center", color="black")

ax.set_title("Potential Impact of Data Science Across Fields")
plt.colorbar(im, orientation='vertical', label='Impact (%)')
plt.tight_layout()
plt.show()

# Step 3: Bar Chart - Individual Metric Analysis
for metric, values in impact_metrics.items():
    plt.figure(figsize=(8, 5))
    plt.bar(fields, values, color='skyblue')
    plt.title(f'Impact of {metric} Across Fields', fontsize=14)
    plt.xlabel('Fields', fontsize=12)
    plt.ylabel('Impact (%)', fontsize=12)
    plt.grid(True)
    plt.show()
```
